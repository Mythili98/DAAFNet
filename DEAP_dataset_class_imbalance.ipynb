```python
import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
table = []
for file in os.listdir('./datasets/Coherence_PLI/'):
    if file.endswith('.mat'):
        data = sio.loadmat('./datasets/Coherence_PLI/' + file)
        data = data['labels_kfold']
        subl = np.zeros(data.shape[0])
        subh = np.zeros(data.shape[0])
        data1 = data[:,0]
        subl[np.where(data1>5)[0]] = 1
        ch1 = np.count_nonzero(subl)
        cl1 = data1.shape[0] - ch1
        data2 = data[:,1]
        subh[np.where(data2>5)[0]] = 1
        ch2 = np.count_nonzero(subh)
        cl2 = data2.shape[0] - ch2
        rows = [file[0:3], ch1, cl1, ch2, cl2]
        table.append(rows)

df = pd.DataFrame(table, columns=['Subject', 'Low_Valence', 'High_Valence', 'Low_Arousal', 'High_Arousal'])
df
```

```python
df = pd.DataFrame(table, columns=['Subjects', 'Low_Valence', 'High_Valence', 'Low_Arousal', 'High_Arousal'])
df1 = df.sort_values(by = 'Subjects')
df1
```

```python
df1.plot.bar(x='Subjects', y=['Low_Valence','High_Valence','Low_Arousal', 'High_Arousal'], stacked=False, figsize=(8,6))
plt.ylabel('Trail count per subject')
plt.title('Subject wise data distribution in various classes')
plt.savefig('data_distribution.pdf', format='pdf')
plt.show()
```
